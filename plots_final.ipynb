```python
def __gauss(i, fac):
    return (fac/2)-abs(i-fac/2)

def smooth(data, fac: int):
    out = np.zeros((len(data)-fac,))
    w_sum = 0
    for i in range(fac+1):
        w = __gauss(i, fac)
        out += w * data[i:len(data)+i-fac]
        w_sum += w
    out /= w_sum
    return np.linspace(fac/2, len(data)-fac/2, len(out), endpoint=False), out
```

```python
plt.figure(dpi=128)
for k, v in training.items():
    if k != 'SAC':
        plt.plot(*smooth(v, 25), label=f'_{k}_unsmoothed', color=light_colors[k])
for k, v in training.items():
    if k != 'SAC':
        plt.plot(*smooth(v, 150), label=k, color=colors[k])
    #else:
    #    plt.plot(training['SAC'], label='SAC', color=colors['SAC'])
plt.title('Model Training')
plt.xlabel('Training Episodes')
plt.ylabel('Smoothed Sum of Rewards')
plt.legend()
plt.show()
```

```python
# Heatmap
res = 32
size = 100
heatmap = {
    k: np.zeros((res,res), dtype=np.int32) for k in eval.keys()
}
def pos2tile(pos: np.array):
    if (pos <= -size).any() or (pos >= size).any(): return None
    return res - 1 - (res * ((pos / (2.0*size)) + 0.5)).astype(np.int32)

fig, axs = plt.subplots(nrows=4, ncols=2, dpi=128, figsize=(6,12))
fig.tight_layout()
names_to_coords = {
    'Random Walk': (0,0),
    'SAC': (0,1),
    'QR-DQN': (1,1),
    'PPO': (2,1),
    'DQN': (3,1),
    'Station Seeker': (1,0),
    'Perciatelli': (2,0)
}
axs[3][0].remove()
for k, v in eval.items():
    poses = v['balloon_state'][...,:2].reshape(-1,2)
    for p in poses:
        t = pos2tile(p)
        if t is not None:
            heatmap[k][tuple(t)] += 1
max_all = 1000 #max(v.max() for v in heatmap.values())
print('max:', max_all)
for k, v in eval.items():
    nc = names_to_coords[k]
    ax = axs[nc[0]][nc[1]]
    im = ax.imshow(heatmap[k], extent=(-size,size,-size,size), vmin=0, vmax=0.8*max_all, cmap='magma')
    ax.set_aspect('equal')
    ax.tick_params(top=True, labeltop=True, bottom=False, labelbottom=False)
    ax.set_xticks(ticks=([-size,0,size] if nc[0]==0 else []))
    ax.set_yticks(ticks=([-size,0,size] if nc[1]==0 else []))
    circle = plt.Circle((0, 0), 50, color='w', fill=False, linewidth=0.5, linestyle='--')
    ax.add_patch(circle)
    ax.set_title(k)
#d = make_axes_locatable(axs[1][1])
#cax = d.append_axes('left', size='5%', pad=0.0)
#fig.colorbar(im, cax=cax, orientation='vertical')
fig.show()
```

```python
steps_3hr = 3 * 60 // 3  # 3 hrs * 60 mins/hr / 3 mins/step = 60 steps
min_energy, max_energy, bin_size = 300, 1410, 30

plt.figure(dpi=128)
for method, _data in data.items():
    power_load = _data['balloon_state'][...,2].flatten()
    # Partition into 3-hr parts and take mean
    power_load = power_load.reshape(-1, steps_3hr).mean(-1)
    # Scale by 3 hrs to convert power to energy
    energy = 3 * power_load
    print(energy)
    # Compute bin counts
    bins_counts = np.array([((energy >= edge) & (energy < edge+bin_size)).sum() for edge in range(min_energy, max_energy, bin_size)])
    bin_percents = 100 * bins_counts / len(energy)
    # Get edge positions (+rand just makes it look nicer visually)
    edges = np.arange(min_energy, max_energy+1, bin_size) + np.random.random()*(max_energy-min_energy)*0.001
    edges[0], edges[-1] = min_energy, max_energy
    plt.stairs(bin_percents, edges=edges,
               linewidth=1.8, label=method, color=colors[method])
plt.title('Energy Consumption')
plt.xlabel('Energy (J)')
plt.ylabel('Frequency (%)')
plt.xlim(min_energy, max_energy)
#plt.yticks(ticks=[0,7])
plt.legend()
plt.show()
```
